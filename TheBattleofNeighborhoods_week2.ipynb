```python
radius = 500
LIMIT = 100

venues = []

for lat, long, borough, neighborhood in zip(final_df['Latitude'], final_df['Longitude'], final_df['Borough'], final_df['Neighborhood']):
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    for venue in results:
        venues.append(( 
            borough,
            neighborhood,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
# convert the venues list into a new DataFrame
venues_df = pd.DataFrame(venues)

# define the column names
venues_df.columns = ['Borough', 'Neighborhood', 'BoroughLatitude', 'BoroughLongitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']

print(venues_df.shape)
venues_df.head()
```

Analyze each area

```python
# one hot encoding
mh_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add postal, borough and neighborhood column back to dataframe
mh_onehot['Borough'] = venues_df['Borough'] 
mh_onehot['Neighborhoods'] = venues_df['Neighborhood'] 

# move postal, borough and neighborhood column to the first column
fixed_columns = list(mh_onehot.columns[-2:]) + list(mh_onehot.columns[:-2])
mh_onehot = mh_onehot[fixed_columns]

print(mh_onehot.shape)
mh_onehot.tail(10)
```

<h3>3.5 Grouped rows by neighborhood and by taking the mean of the frequency of occurrence of each category</h3>

```python
mh_grouped = mh_onehot.groupby(["Borough", "Neighborhoods"]).mean().reset_index()

print(mh_grouped.shape)
mh_grouped
```

<h3>3.6 Created a new dataframe and displaying the top 10 venues for each Neighborhood.</h3>

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
areaColumns = ['Borough', 'Neighborhoods']
freqColumns = []
for ind in np.arange(num_top_venues):
    try:
        freqColumns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        freqColumns.append('{}th Most Common Venue'.format(ind+1))
columns = areaColumns+freqColumns

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Borough'] = mh_grouped['Borough']
neighborhoods_venues_sorted['Neighborhoods'] = mh_grouped['Neighborhoods']

for ind in np.arange(mh_grouped.shape[0]):
    row_categories = mh_grouped.iloc[ind, :].iloc[2:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    neighborhoods_venues_sorted.iloc[ind, 2:] = row_categories_sorted.index.values[0:num_top_venues]

# neighborhoods_venues_sorted.sort_values(freqColumns, inplace=True)
print(neighborhoods_venues_sorted.shape)
neighborhoods_venues_sorted
```

<h3>3.7 Clustering</h3>

```python
# set number of clusters
kclusters = 5

mh_grouped_clustering = mh_grouped.drop(["Borough", "Neighborhoods"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(mh_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.
mh_merged = final_df.copy()

# add clustering labels
mh_merged["Cluster Labels"] = kmeans.labels_

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
mh_merged = mh_merged.join(neighborhoods_venues_sorted.drop(["Borough"], 1).set_index("Neighborhoods"), on="Neighborhood")

print(mh_merged.shape)
mh_merged.head() # check the last columns!
```

```python
# sort the results by Cluster Labels
print(mh_merged.shape)
mh_merged.sort_values(["Cluster Labels"], inplace=True)
mh_merged
```
